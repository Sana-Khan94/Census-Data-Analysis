**We have 117 missing values in Area and Population Column**

Now dealing with missing values since the data type is float so we are taking most frequent value

```python
df['Area\n (In sq. km)'].isnull().sum()
```

```python
df['Area\n (In sq. km)'] = df['Area\n (In sq. km)'].fillna(df['Area\n (In sq. km)'].mode()[0])
```

```python
df['Area\n (In sq. km)'].isnull().sum()
```

```python
df['Population per sq. km.'].isnull().sum()
```

```python
df['Population per sq. km.'] = df['Population per sq. km.'].fillna(df['Area\n (In sq. km)'].mode()[0])
```

```python
df['Population per sq. km.'].isnull().sum()
```

Now their is no missing values left in our data
